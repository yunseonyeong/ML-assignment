## define the feature functions

- feature vector is defined by $(1, f_1(x, y), f_2(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_feature(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    x  = point[:, 0]
    y  = point[:, 1]
    feature = np.array([np.power(y,1), np.power(x,0), np.power(x,1), np.power(x,2), np.power(x,3), np.power(x,4)])

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

## define the linear regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature = $(1, f_1(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_linear_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = np.dot(feature.T, theta.T)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return value
```

## define sigmoid function with input

- $z \in \mathbb{R}$

```python
def sigmoid(z):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = 1 / (1 + np.exp(-z)) 
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value 
```

## define the logistic regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$

```python
def compute_logistic_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    linear = compute_linear_regression(theta, feature)
    value = sigmoid(linear)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define the residual function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_residual(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    logistic = compute_logistic_regression(theta, feature)
    n = logistic.shape[0]
    residual = np.zeros(n)
    for i in range(n):
      residual[i] = (-1)*label[i]*np.log1p(logistic[i]) - (1-label[i])*(np.log1p(1-logistic[i]))

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return residual
```

## define the loss function for the logistic regression

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_loss(theta, feature, label, alpha):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    residual = compute_residual(theta, feature, label)
    n = len(residual)
    residual_sum = np.sum(residual) / n

    norm_theta = np.inner(theta, theta) * (alpha/2)
    
    loss = residual_sum + norm_theta
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return loss
```

## define the gradient of the loss with respect to the model parameter $\theta$

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_gradient(theta, feature, label, alpha):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    logistic = compute_logistic_regression(theta,feature).T
    norm_theta = np.inner(theta, theta) * (alpha/2)
    gradient = np.dot((logistic - label), feature.T) / len(label) + (alpha * norm_theta)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return gradient
```

## compute the accuracy of the prediction for point with a given model parameter

```python
def compute_accuracy(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    num_data = feature.shape[1]
    correct = np.array([None] * num_data)
    
    logistic = compute_logistic_regression(theta, feature).flatten()
    
    is_label_0 = np.array(logistic < 1/2)
    is_label_1 = np.array(logistic >= 1/2)
    
    correct[is_label_0] = 0
    correct[is_label_1] = 1
    
    correct_num = np.array(correct == label)
    accuracy = np.sum(correct_num) / num_data

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return accuracy
```

## initialize the gradient descent algorithm

```python
number_iteration    = 50000 # you can change this value as you want 
learning_rate       = 0.6 # you can change this value as you want 
number_feature      = 6 # you can change this value as you want
alpha               = 0.00001 # you can change this value as you want

theta                       = np.zeros(number_feature)
loss_iteration_train        = np.zeros(number_iteration)
loss_iteration_test         = np.zeros(number_iteration)
accuracy_iteration_train    = np.zeros(number_iteration)
accuracy_iteration_test     = np.zeros(number_iteration)
```

## run the gradient descent algorithm to optimize the loss function with respect to the model parameter

```python
for i in range(number_iteration):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    feature_train = compute_feature(data_train_point)
    feature_test = compute_feature(data_test_point)
    grad = compute_gradient(theta, feature_train, data_train_label, alpha)
    theta           = theta - (learning_rate * grad)
    loss_train      = compute_loss(theta, feature_train, data_train_label, alpha)
    loss_test       = compute_loss(theta, feature_test, data_test_label, alpha)
    accuracy_train  = compute_accuracy(theta, feature_train, data_train_label)
    accuracy_test   = compute_accuracy(theta, feature_test, data_test_label)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    loss_iteration_train[i]     = loss_train
    loss_iteration_test[i]      = loss_test
    accuracy_iteration_train[i] = accuracy_train
    accuracy_iteration_test[i]  = accuracy_test

theta_optimal = theta
```
